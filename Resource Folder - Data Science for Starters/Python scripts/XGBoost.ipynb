```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python

# Define the XGBoost model
xg_reg = xgb.XGBRegressor(objective ='reg:squarederror', colsample_bytree = 0.5, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 300)
```

```python
# Train the model
xg_reg.fit(X,y)

# Make predictions on the test set
y_pred = xg_reg.predict(X)

# Calculate the root mean squared error (RMSE) of the predictions
rmse = np.sqrt(mean_squared_error(y, y_pred))
print("XGBoost: %f" % (rmse))
```

```python
# Train the model
xg_reg.fit(X_train,y_train)

# Make predictions on the test set
y_pred = xg_reg.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("XGBoost: %f" % (rmse))
```

```python
parameters = {'colsample_bytree' : [0.3,0.5],
               'learning_rate' : [0.1,0.2],
                'n_estimators' : [100,300],#,400],
                'max_depth'    : [3,5]
                }
gbr = xgb.XGBRegressor(objective ='reg:squarederror')
cv = RepeatedKFold(n_splits=10, n_repeats=1)#, random_state=1)
grid_search = GridSearchCV(gbr, parameters, cv=cv, n_jobs=-1,verbose=3)#, refit=False)
grid_search.fit(X, y)
print(grid_search.best_estimator_)
print(grid_search.best_score_)
print(grid_search.best_params_)
```
