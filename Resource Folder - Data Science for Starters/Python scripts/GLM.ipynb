```python
import statsmodels.api as sm
import numpy as np
import pandas as pd
```

# POISSON REGRESSION

```python
#import sample data from Poisson_shipdamage csv
df=pd.read_csv(r'C:\Users\ayesha.amjad\Desktop\Analytics for Centerprise\Sample Data\GLM Datasets\Poisson_shipdamage.csv')
display(df.head(5))
```

```python
#define damage as the target variable
target="damage"
```

```python
#set the dataframes for X and y 
X=df.drop(columns=[target])
y=df[target]
```

```python
y
```

```python
#apply one-hot encodings on X
X_onehot=pd.get_dummies(X)
X_onehot
```

```python
# Fit a GLM with a Poisson family distribution
glm_model = sm.GLM(y, sm.add_constant(X_onehot), family=sm.families.Poisson())
glm_results = glm_model.fit()
print(glm_results.summary())
```

# BINOMIAL - LOGIT

```python
#import sample data from Binomial_insurance csv
df2=pd.read_csv(r'C:\Users\ayesha.amjad\Desktop\Analytics for Centerprise\Sample Data\GLM Datasets\Binomial_insurance.csv')
display(df2.head(5))
```

```python
target2='ins'
```

```python
X2=df2.drop(columns=[target2])
y2=df2[target2]
```

```python
# Create an instance of the Logit class for the link function
logit_link = sm.genmod.families.links.Logit()
```

```python
# Fit the GLM with a binomial family distribution and logit link function (without intercept)
glm_binom_logit = sm.GLM(y2, X2, family=sm.families.Binomial(link=logit_link))
glm_results = glm_binom_logit.fit()
print(glm_results.summary())
```

```python
# Fit the GLM with a binomial family distribution and logit link function (with intercept)
glm_binom_logit = sm.GLM(y2, sm.add_constant(X2), family=sm.families.Binomial(link=logit_link))
glm_results = glm_binom_logit.fit()
print(glm_results.summary())
```
